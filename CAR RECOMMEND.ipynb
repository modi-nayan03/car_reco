```python
import ast

columns_to_list = ["carBrand", "carModel", "fuelType", "transmissionType", "bodyType", "city"]

for col in columns_to_list:
    car.loc[:, col] = car[col].astype(str).apply(lambda x: ast.literal_eval(x))
```

```python
car
```

```python
car["tags"] = (
    car["carBrand"] + " " +
    car["carModel"] + " " +
    car["fuelType"] + " " +
    car["transmissionType"] + " " +
    car["bodyType"] + " " +
    car["city"]
)
```

```python
new_df = car[['_id','carBrand','carModel','tags']]
```
